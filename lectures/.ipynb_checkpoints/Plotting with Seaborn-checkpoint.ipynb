# Import scientific packages 

```python
import numpy as np
import pandas as pd
```

# Import and set plotting features

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import matplotlib.pyplot as plt
import seaborn as sns

# making plots look pretty
plt.rcParams['figure.figsize'] = [6.0, 4.0]
plt.rcParams['axes.titlesize'] = 12
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['grid.linestyle'] = '-'
plt.rcParams['legend.fontsize'] = 8
colors = [i['color'] for i in plt.rcParams['axes.prop_cycle']]

sns.set_style("ticks")
```

# Plot

```python
data = pd.read_csv('./data/dataframe.csv')
data.head()
```

```python

sns.distplot(data["output"])

plt.savefig('./figures/figure1.png', dpi=200,bbox_inches='tight')
plt.show()    
```

To mess around with plotting features, it's easier if you make the plot an object (`ax`) and set different features (for example, `ax.grid`, `ax.set_xlim`, `ax.set_yscale`) to the object.

```python
fig, ax = plt.subplots(1, 1)
ax = sns.barplot(data=data, x="Sample", y="output", hue="Sex")
ax.grid(b=True, which='major')
plt.savefig('./figures/figure2.png', dpi=200,bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
sns.despine()
ax = sns.lineplot(data=data, x="output2", y="output", hue="Sex")
ax.grid(b=True, which='major')
ax.set_xlim(0, 15)
ax.set_xlabel("Days", fontweight='bold')
ax.set_yscale("log")
plt.savefig('./figures/figure3.png', dpi=200,bbox_inches='tight')
plt.show()
```
