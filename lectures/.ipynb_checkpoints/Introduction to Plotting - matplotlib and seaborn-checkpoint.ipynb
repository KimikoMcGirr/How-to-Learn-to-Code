# Introduction

Plotting in python can be challenging since there are so many datatypes and so many plotting aspects that you can personalize. In this lesson, I'll cover the basics of the   ```matplotlib.pyplot ``` and the ```seaborn``` plotting packages, give some settings I always add to my notebooks, and show some examples.  

# Importing matplotlib

Calling ```%matplotlib inline``` in the first line of your cell allows matplotlib graphics to show up inline of your notebook. You only need to call it once, and standard practice is to add it into your code when importing your packages. I'll also use the ```seaborn``` and ```pandas``` packages in this lesson.

````python
plt.figure(figsize=(16, 12))
plt.subplot(3, 1, 1)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.ylabel("Y-label")
plt.title("This is a scatter plot")

plt.subplot(3, 1, 2)
data = np.random.rand(50)
plt.boxplot(data)
plt.ylabel("Y-label")
plt.title("This is a boxplot")

plt.subplot(3, 1, 3)
data = np.random.rand(10000)
plt.hist(data, bins=50, facecolor='green')
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("This is a histogram")
````
